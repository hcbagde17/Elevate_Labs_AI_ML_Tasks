```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('melb_data.csv')
df.head(15)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Since more than 50% data is missing in 'BuildingArea', we will drop this column
df = df.drop(columns=['BuildingArea'])
```

```python
# Car has very few missing values, thus replacing with 0
df['Car'] = df['Car'].fillna(0)
```

```python
df['CouncilArea'].value_counts()
```

```python
council_mode = df['CouncilArea'].mode()[0]
df['CouncilArea'] = df['CouncilArea'].fillna(council_mode)
```

```python
# Since 'YearBuilt' has many missing values, but also is an important feature 
# we will fill missing values with median and flag them using new feature
df['YearBuilt_missing'] = df['YearBuilt'].isnull().astype(int)
df['YearBuilt'].fillna(df['YearBuilt'].median(), inplace=True)
```

```python
# cleaned data check
df.isnull().sum()
```

```python
clean_data = df.copy()
clean_data.to_csv('melb_data_cleaned.csv', index=False)
```
