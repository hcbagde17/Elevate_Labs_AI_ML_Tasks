```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
pd.set_option('display.max_columns', None)
```

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
num_cols = df.select_dtypes(include=np.number).columns.tolist()
cat_cols = df.select_dtypes(include='object').columns.tolist()
print("Numerical Columns:", num_cols)
print("Categorical Columns:", cat_cols)
```

```python
for col in cat_cols:
    print(f"Value counts for {col}:")
    print(df[col].value_counts())
    print("\n")
```

```python
df.drop(columns=['education'], inplace=True)  # Dropping 'education' as 'education-num' is present
```

```python
df['workclass'].replace('?', 'other', inplace=True)
```

```python
df['occupation']=df['occupation'].map(lambda x: 'other' if x == '?' else x)
```

```python
df['native.country'] = df['native.country'].apply(
    lambda x: x if x == 'United-States' else 'Other'
)
```

```python
cat_cols.remove('education')
```

```python
for col in cat_cols:
    print(f"Value counts for {col}:")
    print(df[col].value_counts())
    print("\n")
```

```python
df = pd.get_dummies(df, columns=['native.country','workclass', 'marital.status', 'occupation','relationship', 'race'], drop_first=True)
```

```python
df['sex'] = df['sex'].map({'Male': 1, 'Female': 0})
df['income'] = df['income'].map({'<=50K': 0, '>50K': 1})
```
